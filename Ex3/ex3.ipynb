```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#1) Create a simple plot with X and Y axes. Generate random number using linespace method from numpy.
x = np.linspace(0, 10, 50)
y = np.random.rand(50)
plt.plot(x, y)
plt.xlabel('X'), plt.ylabel('Y')
plt.show()
```

```python
#2) Add the elements such as X-axis,Y-axis,X-limits,Y-limits and title for the above graph which you have generated.
x = np.linspace(0, 10, 50)
y = np.random.rand(50)

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Data Plot')

plt.xlim(0, 10)
plt.ylim(0, 1)

plt.grid(True)
plt.show()
```

```python
# 2) Create 2X2 subplots and draw the following in the respective subplots 
# (1) Plot the data points where the KM range from 20000 to 40000 in Toyota.csv 
# (2) Plot the relationship between Age and KM. 
# (3) plot the quantity (counts) based on fuel type 
# (4) Plot the summary of the age attribute from Toyota.csv dataset.
df = pd.read_csv('Toyota.csv')

fig, axs = plt.subplots(2, 2, figsize=(15, 15))
fig.suptitle('Toyota Dataset Analysis', fontsize=16)

# 1. Plot the data points where the KM range from 20000 to 40000
km_range = df[(df['KM'] >= 20000) & (df['KM'] <= 40000)]
axs[0, 0].scatter(km_range['KM'], km_range['Price'])
axs[0, 0].set_title('KM vs Price (20000-40000 KM)')
axs[0, 0].set_xlabel('KM')
axs[0, 0].set_ylabel('Price')

# 2. Plot the relationship between Age and KM
axs[0, 1].scatter(df['Age_08_04'], df['KM'])
axs[0, 1].set_title('Age vs KM')
axs[0, 1].set_xlabel('Age')
axs[0, 1].set_ylabel('KM')

# 3. Plot the quantity (counts) based on fuel type
fuel_counts = df['Fuel_Type'].value_counts()
axs[1, 0].plot(fuel_counts.index, fuel_counts.values, 'go-')
axs[1, 0].set_title('Count by Fuel Type')
axs[1, 0].set_xlabel('Fuel Type')
axs[1, 0].set_ylabel('Count')
axs[1, 0].tick_params(axis='x', rotation=45)

# 4. Plot the summary of the age attribute
age_summary = df['Age_08_04'].describe()
axs[1, 1].plot(age_summary.index, age_summary.values, 'mo-')
axs[1, 1].set_title('Age Summary')
axs[1, 1].set_xlabel('Statistic')
axs[1, 1].set_ylabel('Value')
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
#4) Draw a bar chart with FuelType and its count.
fuel_counts = df['Fuel_Type'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(fuel_counts.index, fuel_counts.values)

plt.title('Count of Vehicles by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')


plt.xticks(rotation=45, ha='right')

for i, v in enumerate(fuel_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.tight_layout()

plt.show()
```

```python
# 5) Draw a horizontal bar chart for FuelType and Price
fuel_price = df.groupby('Fuel_Type')['Price'].mean().sort_values(ascending=True)

plt.figure(figsize=(10, 6))
fuel_price.plot(kind='barh')

plt.title('Average Price by Fuel Type')
plt.xlabel('Average Price')
plt.ylabel('Fuel Type')

for i, v in enumerate(fuel_price):
    plt.text(v, i, f' {v:,.0f}', va='center')

plt.tight_layout()
plt.show()
```

```python
# 6) Compute new_price, include as new column and draw a stacked bar chart for price and new_price against the Age
df['new_price'] = df['Price'] * 1.1

fig, ax = plt.subplots()

ax.bar(df['Age_08_04'], df['Price'], label='Price')
ax.bar(df['Age_08_04'], df['new_price'], bottom=df['Price'], label='New Price')

ax.set_xlabel('Age')
ax.set_ylabel('Price')
ax.set_title('Stacked Bar Chart of Price and New Price against Age')
ax.legend()

plt.show()
```

```python
# 7) Draw the dot plot between car name and age
plt.figure(figsize=(10, 6))
plt.scatter(df['Model'], df['Age_08_04'], alpha=0.6)

plt.xlabel('Car Name')
plt.ylabel('Age')
plt.title('Dot Plot between Car Name and Age')

plt.xticks(rotation=90)

plt.show()
```

```python
# 8) Draw a pie chart with Doors and its count
door_counts = df['Doors'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(door_counts, labels=door_counts.index, autopct='%1.1f%%', startangle=140)

plt.title('Pie Chart of Doors Count')

plt.show()
```

```python
# 9) Draw the box plot to show the statistical summary of Age 
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age_08_04'])

plt.ylabel('Age')
plt.title('Box Plot of Age')

plt.show()
```

```python
# 10) Draw the histogram plot for Age column 
plt.figure(figsize=(10, 6))
plt.hist(df['Age_08_04'], bins=10, edgecolor='black', alpha=0.7)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

plt.show()
```
